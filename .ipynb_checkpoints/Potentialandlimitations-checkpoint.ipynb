```python
import numpy as np

def random_qubit_density_bloch():
    # random direction
    v = np.random.normal(size=3)
    v = v / np.linalg.norm(v)
    
    # random radius in [0,1]
    r = np.random.rand()
    r_vec = r * v
    
    rx, ry, rz = r_vec
    
    rho = 0.5 * np.array([
        [1 + rz, rx - 1j*ry],
        [rx + 1j*ry, 1 - rz]
    ])
    
    return rho
```

```python
matrices = np.zeros((1000, 2, 2), dtype=complex)
```

```python
for i in range(1000):
    matrices[i] = random_qubit_density_bloch()
```

```python
for i in range(10):
    print(matrices[i])
```

```python

I = np.array([[1, 0],
              [0, 1]], dtype=complex)

X = np.array([[0, 1],
              [1, 0]], dtype=complex)

Y = np.array([[0, -1j],
              [1j,  0]], dtype=complex)

Z = np.array([[1,  0],
              [0, -1]], dtype=complex)

def expectation_pauli_x(rho):
    return np.array(np.trace(rho @ X).real)
```

```python
exp_x = expectation_pauli_x(matrices)
```

```python
for i in range(10):
    print(exp_x[i])
```
