```python
import numpy as np
import qutip as qt

#testing function to compute partial trace

#firt of all let's define two matrices and consider their tensor product
#the first matrix is 2x2
A = np.array([[1, 2], [3, 4]])
A = A + A.conj().T
A = A / np.trace(A)
print(A)
#the second matrix is 3x3
B = np.array([[5, 6, 7], [8, 9, 10], [11, 12, 13]])
B = B + B.conj().T
B= B / np.trace(B)
print(B)
#their tensor product is a 6x6 matrix
C = np.kron(A, B)
print("Tensor product of A and B:")
print(C)

A_red = myQML.partial_trace(C,[2,3],0)
print("Partial trace of C over the second subsystem:")
print(A_red)

B_red = myQML.partial_trace(C,[2,3],1)
print("Partial trace of C over the first subsystem:")
print(B_red)
C_Qutip = qt.Qobj(C, dims=[[2, 3], [2, 3]])
A_red_qutip = qt.ptrace(C_Qutip, 0)
print("Partial trace of C over the second subsystem using qutip:")
print(A_red_qutip)
B_red_qutip = qt.ptrace(C_Qutip, 1)
print("Partial trace of C over the first subsystem using qutip:")
print(B_red_qutip)
```

```python
print(np.allclose(A_red, A))
print(np.allclose(B_red, B))
```
