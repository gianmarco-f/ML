```python
import numpy as np
```

```python


N0 = 3 # input neurons
N1 = 2 # output neurons

w = np.random.randn(N1,N0) # Gaussian random numbers in N1 x N0 array
b = np.random.randn(N1) # Gaussian random numbers in N1 vector

y_in = np.random.randn(N0) # input vector

z = np.dot(w,y_in) + b # linear transformation
y_out = 1/(1+np.exp(-z)) # sigmoid activaction function
```

```python
print(y_out)
```
